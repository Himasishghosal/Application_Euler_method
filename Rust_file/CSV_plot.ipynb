```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv('/content/euler_comparison.csv')

# Plot error vs. time
plt.figure(figsize=(8, 5))
plt.plot(data['t'], data['error'], marker='.', color='red', label='|Exact - Numerical|')
plt.xlabel('Time t')
plt.ylabel('Error')
plt.title('Error Between Numerical and Exact Solution (Euler Method)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("euler_error_plot.png")
plt.show()
```
